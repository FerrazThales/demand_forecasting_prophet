Uma das principais tarefas de um **Engenheiro de Produção** é otimizar *recursos*. Atualmente um dos recursos mais *valiosos* e *rentáveis* de uma organização são os **dados**, afinal [data is the new oil](https://www.forbes.com/sites/forbestechcouncil/2019/11/15/data-is-the-new-oil-and-thats-a-good-thing/?sh=32998a0c7304). Podemos utilizar estes dados para **projetar**, ou seja, antecipar a *demanda* de diversos tipos de produtos e assegurar a **eficiência** da *gestão* de toda nossa cadeia de suprimentos. Por isso, uma **previsão de demanda** bem feita pode gerar *vantagem competitiva* nas decisões do negócio e **reduzir** absurdamente os *custos*.

É **imprescindível** que as empresas entendam qual será a demanda **futura** para que seus gestores estejam aptos a tomarem *decisões* mais **assertivas** sobre os **preços**, potencial de **mercado**, compra de **matéria-prima** e até mesmo, possibilidade de **extensão** do negócio. Além disso, a previsão de demanda é essencial para a **Gestão do Estoque**, uma vez que projeções *inadequadas* podem gerar *acúmulo* ou *falta* de **produtos** - e tudo isso tem um gigantesco impacto financeiro negativo nas organizações.

Ao se fazer uma *projeção de demanda* **eficaz** as empresas adquirem inúmeras **vantagens** práticas, como o aumento da **produtividade** dos colaboradores e a **diminuição** do trabalho *ocioso* - isto acontece porque os processos se tornam mais **otimizados**. Além disso, todo o planejamento relacionado as atividades de  orçamento, marketing e estratégias de preço são melhor estruturados, já que existem mais **informações** disponíveis. É importante ressaltar que sem uma previsão de demanda **adequada** o negócio assume um grande **risco** de operar e produzir um produto que o **mercado** não precisa.

Tradicionalmente, a *previsão de demanda* de divide em **duas** grandes *categorias*. As projeções baseadas em termos **qualitativos** e **quantitativos**. As metodologias *qualitativas* são mais **subjetivas** e se baseiam em pesquisas com especialistas da área, sempre buscando um consenso. Um grande exemplo dos métodos qualitativos é o [Método Delphi](https://www.questionpro.com/blog/pt-br/metodo-delphi/). 

Já as metodologias *quantitativas* se segmentam em outras *duas* **subcategorias**. A primeira se orienta por **métodos estatísticos** e utiliza operações *matemáticas* para realizar suas previsões. Um modelo simples que aplica este método é o das [Médias Móveis](http://www.fecilcam.br/anais/x_eepa/data/uploads/1-engenharia-de-operacoes-e-processos-da-producao/1-02.pdf). Já a segunda subcategoria se baseia em **algoritmos** de machine learning (*até mesmo deep learning*), ou seja, o *aprendizado* do modelo é feito através dos dados com uma ajuda **computacional**. Um grande exemplo de método utilizado nesta subcategoria é a [Arquitetura LTSM](https://machinelearningmastery.com/how-to-develop-lstm-models-for-time-series-forecasting/).

Desta forma, com o avanço da **tecnologia** e das oscilações de mercado é muito importante ter uma previsão de demanda **robusta** e que contempla seu modelo de negócio, uma vez que este tipo de **ferramenta** já se tornou uma vantagem competitiva em todos os mercados.





#O Mercado de Vinhos e suas tendências.

<center><img src="https://image.freepik.com/fotos-gratis/copo-de-vinho-tinto-e-garrafa-no-balcao-de-bar_107420-65845.jpg" width="70%"></center>

O vinho é uma bebida que **acompanhou** a história das civilizações. Ele esteve presente nas culturas egípcias, gregas, romanas e até mesmo apareceu no *antigo testamento* sendo consumido pelo povo Judeu. [A sua origem ainda é incerta](https://blog.famigliavalduga.com.br/a-historia-do-vinho-no-mundo-entenda-como-esta-bebida-colonizou-os-continentes/), mas provavelmente ele surgiu nas áreas mediterrâneas quando os seres humanos abandonaram o nomadismo.

A expansão desta bebida **fermentada** pelo mundo ocorreu principalmente pelo povo *romano*. Eles inclusive *classificaram* e *catalogaram* diversas variedades de uvas e **pragas** que poderiam atingir as parreiras. Além disso, nesta época desenvolveram o barril e colocaram os vinhos em garrafas de vidro.

Outra cultura que ajudou a *expansão* desta bebida foi seu uso na **eucaristia**. Além da passagem de **transformação** de água em vinho, Jesus em sua última refeição declarou: "*Este é o meu sangue*". Portanto, o vinho acabou estando [intimamente relacionado a cultura dos cristãos](https://revistaadega.uol.com.br/artigo/o-significado-do-vinho-na-religiao_5342.html). Um fato **curioso** é que o vinho utilizado na celebração das missas não podem ser qualquer um, é necessário que a bebida utilizada siga as **especificações** do Concílio de Trento.

A [história do vinho no Brasil](https://www.enologia.org.br/curiosidade/a-historia-do-vinho-no-brasil) começa no século XVII com **jesuítas** trazendo videiras da Europa para a utilização em **cultos religiosos**. No entanto, as variedades trazidas não se **adaptaram** muito bem ao *clima* e solo *brasileiro*. Por conta disso, a **disseminação** do cultivo de uva **não** *prosperou* em nossas terras.

Apesar disso, no século XIX houve um grande **crescimento** da *vitivinicultura* em nosso país, principalmente devido a chegada de **italianos**. Embora as variedades que trouxeram ainda não se **adaptassem** muito bem ao solo brasileiro, foi a *garra* e *determinação* do povo italiano que mantiveram os hábitos do **consumo** de vinho deste povo ativos. Desta forma, eles **buscaram** incessantemente em todo o mundo *espécies* que pudessem **cultivar** em nosso país para conseguir manter sua práticas.

Já no século XX, mais precisamente na década de 70, as vinícolas **multinacionais** aportaram no Brasil, se estabelecendo principalmente no *Rio Grande do Sul*. Estas empresas trouxeram uma grande preocupação com a **qualidade** e técnicas de cultivo das videiras. Graças a isso, a produção no Brasil passou por um processo de **modernização** e hoje os vinhos *brasileiros* são considerados com um padrão de qualidade **internacional**.

Mas infelizmente nosso país ainda enfrenta alguns **desafios** para consolidar o hábito de consumo de **vinho** pelas pessoas. O clima **quente** desfavorece o consumo deste tipo de bebidas, há uma *dificuldade* de **importação** das principais marcas e os **preços** são considerados altos se comparados com outros tipo de **bebidas** alcoólicas.

No entanto, [o consumo de vinho aumentou em 2020 no Brasil](https://g1.globo.com/economia/noticia/2021/04/20/consumo-de-vinho-tem-alta-de-18percent-em-2020-puxado-pelo-e-commerce.ghtml) principalmente devido ao **e-commerce**. De acordo com o site [Winext](https://www.wine-xt.com/pt-br/blog/2021/2/3/oportunidades-crescimento-mercado-de-vinho-2021)esta maior procura em ambientes **digitais** são explicadas porque os consumidores podem **facilmente** explorar, descobrir, *comparar* e *comentar* produtos, o que torna o canal particularmente **atrativo** para a categoria do vinho e a sua enorme **diversidade** de tipos, estilos, origens e marcas.

Como mostrado na imagem abaixo , cerca de 59% dos *brasileiros* consideram comprar vinho de forma **online** no futuro. Este maior engajamento com os mercados digitais vão obrigar as produtoras destas bebidas alcoólicas a se **modernizarem**, uma vez que terão de apresentar sites **atrativos**, um banco de dados **robusto**, recomendadores de produtos **eficazes** e tudo mais que a presença digital necessita. Além disso, para entregas **rápidas** e de **baixo** custo será necessário um grande empenho *logístico*, principalmente na estruturação de **Centros de Distribuição** e na importação das marcas tradicionais de vinho.

Além disso, a variável `item_id` **não** precisa ter a formatação *númerica*. Uma vez que não faz sentido falar na *média* dos `item_id` e nem será feita operação matemática com esta variável.

## Juntando as duas bases de dados

Ok, eu já devo ter dito várias vezes que temos duas bases de dados. Portanto, esta será a última: *temos duas bases de dados separadas*. Precisamos agrupá-las de uma maneira que faça sentido para o nosso projeto, agrupando o que é comum.

Existem diferentes estratégias para agrupar dataframes, este [artigo do Data Hackers](https://medium.com/data-hackers/pandas-combinando-data-frames-com-merge-e-concat-10e7d07ca5ec) explora muito bem esta questão. Inclusive a imagem abaixo foi retirada de lá.

Entenda o *x* do desenho como nosso **catálogo** de produtos e o *y* como a base de dados de **vendas**. Nosso objetivo é **prever** a quantidade de produtos que serão vendidos nos próximos **quatro** meses. Por isso, precisamos manter todos os valores do conjunto *y*. É nele que contém estes valores sobre as quantidades vendidas.

Por outro lado, precisamos do conjunto *x* para nos dizer **QUAL** produto foi vendido. E isso o **código** (`item_id`) do conjunto *x* explica, porque ele é igual tanto para o catálogo quanto para as vendas.

Avaliando os parágrafos acima, parece uma boa **escolher** o `how` como *right*. Uma vez que queremos tudo referente a base de dados de **vendas** e somente o **necessário** do **catálogo** de produtos.

```
#códigos para juntar as duas bases de dados
df = pd.merge(produtos,vendas,how='right',left_on='item_id',right_on='item')

#mudando a variável data para o índice
df.set_index('date',inplace=True)

#retirando a coluna item duplicada
df.drop('item',axis=1,inplace=True)
```

Após **agrupadas** as duas bases de dados, podemos **aprofundar** nossa análise.

Com os dados de **quantidades** vendidas e o **preço** de cada vinho, é possível calcular a **receita** obtida pela empresa naquela negociação. Por conta disso, precisamos criar **novas** *variáveis* em nossa base de dados que se referem a estas receitas. 

Além disso, por termos a variável `date` é aconselhável **explorarmos** como se comportam os dados na **perspectiva** de um *ano*, na perspectiva semanal,etc. Desta forma, criaremos **novas** variáveis para **cada** atributo de **data** que seja *necessário*.

```
#receitas em dólares
df['revenue_usd'] = df.sales * df.price_usd

#receitas em reais
df['revenue_brl'] = df.sales * df.price_brl

#ano da vendas
df['year'] = df.index.year

#mês das vendas
df['month'] =  df.index.month_name()

#dia das vendas
df['day'] = df.index.day

#dia da semana
df['day_of_week'] = df.index.dayofweek

#nome do dia na semana
df['day_name'] = df.index.day_name()

#final de semana
df['weekend'] = df.day_of_week > 4
```

## Examinando os Outliers

Uma boa maneira de **visualizar** os valores e as *estatísticas descritivas* de nossa base de dados é através de um [Boxplot](https://www.voitto.com.br/blog/artigo/boxplot). No primeiro **Boxplot** (*esquerda*) verificamos a **quantidades** de vinhos vendidos **diariamente**. Em **média**, esta empresa vende 46 mil vinhos ao dia e esta soma resultado numa venda em torno de 16 milhões de garrafas no ano.     Este valor é um pouco **alto** para os padrões brasileiros. Neste momento, uma boa alternativa seria sentar com a área de **negócio** e *repassar* estes valores, se possível de um a um. 

No entanto, para esta *nossa* análise, e considerando que também não existem dados **duplicados**, vamos assumir as entradas como **corretas** e manter nossa base de dados já que não foi apresentado nenhum *outlier* **relevante**.

Já pelo **Boxplot** relacionado aos **preços** dos vinhos (*direita*), observamos que existem **alguns** outliers. No entanto, se [buscarmos os preços destes vinhos](https://www.wine-searcher.com/) na internet, notaremos que estes preços são **plausíveis** aos vendidos no mercado. Portanto, manteremos estes registros em nossa base de dados.

Para uma análise mais apurada dos outliers, utilizamos o método [IQR - Method](https://online.stat.psu.edu/stat200/lesson/3/3.2#:~:text=Any%20observations%20that%20are%20more,above%20Q3%20are%20considered%20outliers.). Com este método calculamos a distância entre o *terceiro* e o *primeiro* quartil e estabelecemos os **limites** superior e inferior. Qualquer observação **acima** do limite superior é considerada um **outlier**, bem como qualquer valor **abaixo** do limite inferior também é considerado um **outlier**.

```
#agrupando a quantidade de vinhos vendidas de forma diária
quantidade_de_vinhos_vendida_por_dia = df.groupby(by='date').sum()

#criando o ax e o fig no matplotlib
fig,ax = plt.subplots(figsize=(16,8),dpi=80,ncols=2)

#criando o plot para receitas em dólar
sns.boxplot(x=quantidade_de_vinhos_vendida_por_dia.sales,palette='Blues',linewidth=0.6,ax=ax[0])
ajuste_layout(ax=ax[0],location='center',
              titulo='Quantidade de Vinhos Vendidos Diariamente',
              xlabel='sales',
              grid=True,grid_lw=0.4)

#criando o plot para o preço do vinho
sns.boxplot(data=df,x='price_usd',palette='Purples',linewidth=0.6,ax=ax[1])
ajuste_layout(ax=ax[1],location='center',
              titulo='Preços dos vinhos em dólar',
              xlabel='price_usd',
              grid=True,grid_lw=0.4)

#plotando o gráfico
plt.tight_layout(pad=4.0);
```

```
#avaliando a porcentagem de no preço do vinho
IQR_sales = quantidade_de_vinhos_vendida_por_dia.sales.quantile(0.75) - quantidade_de_vinhos_vendida_por_dia.sales.quantile(0.25)
lim_sup_sales = quantidade_de_vinhos_vendida_por_dia.sales.quantile(0.75) + (1.5 * IQR_sales)
lim_inf_sales = quantidade_de_vinhos_vendida_por_dia.sales.quantile(0.25) - (1.5 * IQR_sales)

#verificando a porcentagem de outliers na variável price_usd e a receita provindas deles
outliers_sales = quantidade_de_vinhos_vendida_por_dia.sales[
                                                            (quantidade_de_vinhos_vendida_por_dia.sales < lim_inf_sales) | 
                                                            (quantidade_de_vinhos_vendida_por_dia.sales > lim_sup_sales)]
quantidade_outliers_sales = (outliers_sales.shape[0] / quantidade_de_vinhos_vendida_por_dia.shape[0])

#avaliando a porcentagem de no preço do vinho
IQR_price_usd = df.price_usd.quantile(0.75) - df.price_usd.quantile(0.25)
lim_sup_price_usd = df.price_usd.quantile(0.75) + (1.5 * IQR_price_usd)
lim_inf_price_usd = df.price_usd.quantile(0.25) - (1.5 * IQR_price_usd)

#verificando a porcentagem de outliers na variável price_usd e a receita provindas deles
outliers_price_usd = df[(df.price_usd < lim_inf_price_usd) | (df.price_usd > lim_sup_price_usd)]
quantidade_outliers_priced_usd = (outliers_price_usd.shape[0] / df.shape[0])

#criando um dataframe com os valores
pd.DataFrame({'Variável': ['sales','price_usd'],
              'Limite Superior':[lim_sup_sales,lim_sup_price_usd],
              'Fração de Outliers':[quantidade_outliers_sales,quantidade_outliers_priced_usd]}).set_index('Variável').round(2)
```

Como dito anteriormente, esta empresa está vendendo em torno de 16 **milhões** de garrafas por ano e isto, provavelmente, precisaria ser **validado** com os responsáveis pelo negócio. Estes registros estão proporcionando uma receita **anual** de quase *3* **bilhões** de *dólares*, o que automaticamente colocaria a empresa dentre os [grandes players do mercado de distribuição de vinho](https://winesvinesanalytics.com/features/article/202690/Top-10-US-Wine-Distributors).

#Séries Temporais - chegamos nelas! :)

Séries temporais são **observações** feitas sobre uma variável em intervalos **regulares** de tempo. Dependendo da **frequência** em que são realizadas estas observações, ela pode ser chamada de uma série temporal diária, mensal, anual, etc.

É importante ressaltar que como existe esta **dependência** do tempo, as técnicas tradicionais de **regressão** não são adequadas para este tipo de problema. Além disso, para estas análises se faz a [suposição](https://www.inf.ufsc.br/~marcelo.menezes.reis/Cap4.pdf) que existem causas **não aleatórias** que exerceram **influência**  sobre o comportamento dos dados no passado e **podem** contribuir para isso no futuro. 

O objetivo de se analisar qualquer série temporal é **orientar** a tomada de decisão a partir do **conhecimento** dos padrões observados ao longo do tempo e do comportamentos dos dados **históricos**.

Qualquer série temporal pode ser divida em basicamente 3 componentes:

* Tendência: é o crescimento/decrescimento da variável ao longo do tempo observado. Existem alguns [testes para verificar esta tendência](https://operdata.com.br/blog/caracteristicas-das-series-temporais/), como o Teste de Wald e o de Cox-Stuart. 
* Sazonalidade: são padrões que se repetem em intervalos regulares de tempo. Um exemplo, seria uma maior vendas de sorvetes no verão.
* Erros residuais: são flutuações inexplicáveis, ruídos de nossos dados.

**Obs: a série temporal não precisa ter obrigatoriamente tendência, erros ou sazonalidade.**

Se você utiliza apenas os valores **passados** da sua *time series* para realizar suas **previsões**, será utilizado modelos de previsão de séries temporais [univariados](http://www.bertolo.pro.br/MetodosQuantitativos/Previsao/pmc44.htm).

No entanto, se **além** dos dados passados você utiliza variáveis **exógenas** - outras variáveis que variam ao longo do tempo e que tem **relações** entre si, você utilizará modelos de previsão de séries temporais [multivariados](https://www.analyticsvidhya.com/blog/2018/09/multivariate-time-series-guide-forecasting-modeling-python-codes/).

Para este **projeto** será escolhida a variável `sales`, portanto será feita uma previsão **univariada**. A figura abaixo traz os comportamentos dos dados em relação aos intervalos de tempo **escolhidos**.

### Verificando se a série é estacionária

Quando queremos fazer **previsões** em séries temporais um dos **pré-requisitos** é que ela seja **estacionária**. Por conta disso ela deve ter as **propriedades** estatísticas como: a média, variância e covariâncias **constantes** em relação ao tempo.

Um **plot** de uma série estacionária se comportaria de maneira **semelhante** a imagem abaixo.

<center><img src="https://cdn.analyticsvidhya.com/wp-content/uploads/2018/09/ns6-e1536674898419.png" width="35%"></center>



Mas como **verificamos** que uma série temporal é *estacionária*?

Uma maneira muito utilizada é aplicar o [Teste Aumentado de Dick Fuller](https://www.machinelearningplus.com/time-series/augmented-dickey-fuller-test/), também conhecido como teste ADF. 


O ADF é um dos **testes estatísticos** mais usados para Time Series. Suas hipóteses **nula** e alternativa são:

* A Hipótese Nula ($H_0$) do teste é que a Time Series **não é estacionária**. Ou seja, possui algum tipo de dependência em relação ao tempo.
* A Hipótese Alternativa ($H_1$) **rejeita a hipótese nula**, ou seja, que a Time Series é estacionária.

Aplicaremos o ADF em nosso projeto através de um **módulo** do StatsModels, o [adfuller](https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html). Ele retorna o teste estatístico do ADF, o [p-valor](https://www.scielo.br/j/jbpneu/a/SWk5XsCsXTW7GBZq8n7mVMJ/?format=pdf&lang=pt#:~:text=O%20valor%2Dp%20indica%20a,poder%20suficiente%20para%20detect%C3%A1%2Dlo.) e alguns [valores críticos](https://support.minitab.com/en-us/minitab-express/1/help-and-how-to/basic-statistics/inference/supporting-topics/basics/what-is-a-critical-value/#:~:text=In%20hypothesis%20testing%2C%20a%20critical,and%20reject%20the%20null%20hypothesis.) para avaliarmos nosso teste de **hipótese**.

**Mas e se a série temporal não for estacionária? O que devemos fazer?**

Devemos torná-la estacionária através de diferentes **técnicas**. Uma delas pode ser a [diferenciação](https://medium.com/ensina-ai/princ%C3%ADpios-b%C3%A1sicos-para-criar-previs%C3%B5es-de-s%C3%A9ries-temporais-e58c451a25b), que é a **subtração** do valor corrente pelo valor **anterior**. Ou a [transformação logarítmica](https://www.kaggle.com/rdizzl3/time-series-transformations), que é basicamente a aplicação do **logaritmo** natural nos **registros** do dataset. Para se aprofundar um pouco mais nas diferentes técnicas sugiro a **leitura** deste [link](https://people.duke.edu/~rnau/411fcst.htm) da *Duke University*.


```
#extrarir somente os valores
x = agregando_valores_mensais

#aplicar ADF e imprimir os resultados
result = adfuller(x.sales)
print('Teste Dickey–Fuller Aumentado:')
print('Teste estatístico: ','{:.2f}'.format(result[0]))
print('P-valor: ','{:.5f}'.format(result[1]))
print('Valores Críticos:')
for key, value in result[4].items():
	print('\t{}: {:.2f}'.format(key, value));
print(f'\nInferência: A Time Series é {"não-" if result[1] >=0.05 else ""}Estacionária.')
```

A partir do teste ADF podemos **inferir** com um nível de **significância** de quase 2% que nossos dados representam uma série temporal **estacionária**. Desta forma, estamos prontos para realizar as **previsões** de vendas para os próximos 4 meses!

Neste projeto utilizaremos 2 **modelos**: o SARIMA (estatística tradicional) e o Prophet (framework de Time Series do Facebook).

```
# definindo o tamanho da previsão para os quatro meses seguintes
prediction_size = 4
train = agregando_valores_mensais[:-prediction_size]
validation = agregando_valores_mensais[-prediction_size:]


#verificando a quantidade de valores
print('Treino:',train.shape[0],'registros')
print('Validação:',validation.shape[0],'registros')
```

# SARIMA

[ARIMA](https://www.machinelearningplus.com/time-series/arima-model-time-series-forecasting-python/), é um **acrônimo** de *Auto Regressive Integrated Moving Average*. Ele é um **algoritmo** de *previsão* que se baseia na ideia de que valores passados da *Time Series* podem, **sozinhos**, ser utilizados para predizer os valores **futuros**.

Um passo importante para utilizar o **ARIMA** é tornar a Time Series como **estacionária**, uma vez que o termo *Auto Regressive* se refere a um modelo de [regressão linear](https://www.machinelearningplus.com/machine-learning/complete-introduction-linear-regression-r/). Isto é importante *porque* os **modelos** de *regressão linear* funcionam como **melhor** previsores quando não há **correlação** e há **independência** entre as variáveis.

A **ARIMA** é caracterizada pelos seguintes *parâmetros*:

* Auto Regressive - **p** 
  *  Número de lag (períodos) utilizados no modelo auto regressivo.

* Moving Average - **q**
  *  O número utilizado para o cálculo da média móvel, também conhecido como a ordem da média móvel.

* Integrated - **d**
  * Está relacionado a quantidade de **transformações** mínimas realizadas para que a Time Series se **transforme** em *estacionária*. Quando a *Time Series* já é estacionária o valor de `d` se torna **zero**.

O grande **problema** em aplicar o modelo **ARIMA** em nossa análise é que ele não concebe a **sazonalidade**. Por conta disso, como nossa base de dados **possui** *sazonalidade*, devemos utilizar o **SARIMA** -  o **S** não é de sextou e nem de saudade, mas sim de *seasonality*.

Para **facilitar** nosso trabalho, neste projeto utilizaremos uma **api** chamada [auto-arima](https://alkaline-ml.com/pmdarima/modules/generated/pmdarima.arima.auto_arima.html). Ela possibilita descobrir **automaticamente** quais são os parâmetros **ótimos** para o nosso modelo utilizando como critério de escolha o [AIC](https://coolstatsblog.com/2013/08/14/using-aic-to-test-arima-models-2/). Além disso, podemos **setar** o parâmetro `seasonal` como `True` e escolher o período de sazonalidade pelo parâmetro `m` - neste caso foi escolhido **12** por se tratar de resultados **anuais**.


```
#Utilizando o auto_arima que seleciona automaticamente o melhor modelo
sarima_model = auto_arima(y=train.sales,start_p=1,start_q=1,
                          test='adf',m=12,seasonal=True,stepwise=True,trace=True)
```

#Prophet

Os desenvolvedores do **Facebook** perceberam que as técnicas **tradicionais** para realizar *forecasting* eram muito **inflexíveis** e incapazes de incorporar as nuances de cada **negócio**. Além disso, também observaram que era muito **raro** encontrar profissionais **qualificados** que tivessem a capacidade de fazer uma **boa** previsão.

Por conta disso, eles construiram uma **ferramenta** open-source, o [Prophet](https://facebook.github.io/prophet/). Naturalmente, o `Prophet` foi criado baseado na **experiência** do dia-a-dia que os desenvolvedores do *Facebook* enfrentavam.Ou seja, é previsor que considera a existência de **feriados**, observações **ausentes**, uma grande quantidade de **outliers**,etc.

Basicamente o `Prophet` utiliza um [modelo aditivo](http://www.producao.ufrgs.br/arquivos/disciplinas/119_teq6_st_decomposicao.pdf) que avalia a **tendência** (por métodos lineares ou regressão logística), a sazonalidade das **semanas** por [Séries de Fourier](https://www.ime.unicamp.br/sites/default/files/lem/material/series-de-fourier.pdf) e uma sazonalidade **semanal** por variáveis [dummy](https://en.wikipedia.org/wiki/Dummy_variable_(statistics)).

Caso tenha mais **interesse** de avaliar como funciona o `Prophet` mais internamente, dê uma lida no [paper](https://peerj.com/preprints/3190.pdf) escrito por seus **desenvolvedores**.


```
#ajustando a base de dados para utilizar no prophet
train_prophet = train.reset_index()

#renomeando o nome das colunas
train_prophet.columns = ['ds','y']

#visualizando as 5 últimas entradas
train_prophet.tail()
```

```
#instanciando o prophet
model_prophet = Prophet()

#realizando o fit do prophet
model_prophet.fit(train_prophet)

#determinar o periodo para previsão, o freq='M' é porque se tratam de meses
future = model_prophet.make_future_dataframe(periods=prediction_size,freq='M')
```

# Realizando as Previsões

Para calcular os **erros** de nossos modelos usaremos um *módulo* do scikit-learn, o [Mean Squared Error](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html). Setaremos o **parâmetro** `squared` como *False* para calcular a **raiz quadrada** deste valor e **conseguir** comparar com valores que foram *estimados*.

Foi optado esta **métrica** porque queremos **reduzir** ao máximo a *diferença* entre os valores previstos e reais. Além disso, pretendemos utilizar o modelo que *menos* proporciona **outliers**, uma vez que **errar** demasiadamente o valor esperado pode gerar **custos** onerosos para a empresa de vinhos analisada.

A imagem abaixo ilustra como é realizado o cálculo da métrica utilizada, o [RMSE (Root Mean Squared Error)](https://www.statisticshowto.com/probability-and-statistics/regression-analysis/rmse-root-mean-square-error/).

<p align='center'>
<img src="https://miro.medium.com/max/966/1*lqDsPkfXPGen32Uem1PTNg.png" width="30%">
</p>

```
#função para printar o erro
def printando_o_erro(y_true,y_pred,previsor):
  error = mse(y_true=y_true,y_pred=y_pred,squared=False)
  print('A raiz do erro-médio quadrado utilizando o modelo {} é de {:.2f}'.format(previsor,error))
```

Fazendo as *previsões* do **SARIMA** e avaliando o *modelo*.

```
# fazer a previsao com SARIMA
forecasts,conf_int = sarima_model.predict(n_periods=prediction_size,return_conf_int=True,alpha=0.8)

#criando um dataframe com as previsões e os intervalos de confiança
previsoes_sarima = pd.DataFrame({'data':validation.index,
                                'previsoes':forecasts,
                                'lower_conf':conf_int[:,0],
                                'upper_conf':conf_int[:,1]})

#visualizando o dataframe de previsoes
previsoes_sarima
```

```
#criando o fig e o ax no matplotlib
fig,ax = plt.subplots(figsize=(12,8))

#criando o gráfico dos dados de treino
plt.plot(agregando_valores_mensais,color='blue',alpha=0.55,marker='.')
plt.fill_between(data=previsoes_sarima,x='data',y1='lower_conf',y2='upper_conf',color='blue',alpha=0.2)
ajuste_layout(ax=ax,
              titulo='Realizando as previsões para os próximos 4 meses utilizando o SARIMA',
              ylabel='Quantidades Vendidas',
              grid=True,grid_lw=0.6,grid_style='dotted')

#plotando o gráfico
plt.tight_layout();
```

```
#calcular o erro
printando_o_erro(y_true=validation.sales,y_pred=forecasts,previsor='ARIMA')
```

Fazendo as *previsões* do **Prophet** e avaliando o *modelo*.

```
#realizar a previsão com o período mensal estabelecido
forecast = model_prophet.predict(future)

#visualizando os 4 últimos registros
forecast.tail(4)
```

```
#crian do ax e o plt
fig, ax = plt.subplots(figsize=(14,8))

#criando gráfico
model_prophet.plot(forecast,ax=ax)
ajuste_layout(ax=ax,
              titulo='Realizando as previsões para os próximos 4 meses utilizando o Prophet',
              ylabel='Quantidades Vendidas',
              grid=True,grid_lw=1.8,grid_style='dotted')

#plotando os gráficos
plt.tight_layout();
```

```
#calcular o erro
printando_o_erro(y_true=validation.sales,y_pred=forecast[-prediction_size:].yhat,previsor='Prophet')
```
